Trong dữ liệu có nhiều bệnh nhân có thời gian bắt đầu điều trị ICU lần đầu tiên diễn ra trước thời gian nhập viện. Điều này có thể lí giải là nhiều bệnh nhân nhập viện trong tình
trạng khẩn cấp (bị bệnh nặng, tai nạn hoặc thai phụ sinh sản) nên được đưa vào ICU trước khi làm thủ tục nhập viện. Chúng tôi giải quyết những trường hợp này bằng cách thay thời gian nhập viện bởi thời gian vào ICU lần đầu tiên.

```python
# Cập nhật thời gian nhập viện cho những bệnh nhân có intime trước admittime
patients.loc[patients['intime'] > patients['admittime'], 'admittime'] = patients['intime']
```

Có một số bệnh nhân có thời gian nhập viện xảy ra sau thời gian xuất viện. Điều này là vô lý, do đó chúng tôi loại bỏ những bệnh nhân này khỏi bộ dữ liệu.

```python
# Loại bỏ những bệnh nhân có dischtime trước admittime
patients = patients[patients['dischtime'] > patients['admittime']]
```

```python
patients[patients['dischtime'] < patients['admittime']][['subject_id', 'hadm_id', 'icustay_id', 'admittime', 'dischtime']]
```

---

```python
# Tính tổng thời gian nằm viện và lưu vào cột total_los
patients.loc[:, 'total_los'] = (patients['dischtime'] - patients['admittime']).dt.total_seconds() / 86400
```

```python
patients[patients['total_los'] < patients['los_icu']][['subject_id', 'hadm_id', 'icustay_id', 'los_icu', 'total_los']] 
```

```python
patients[patients['total_los']+0.5 <= patients['los_icu']][['subject_id', 'hadm_id', 'icustay_id', 'los_icu', 'total_los']] 
```

Có 4400 bệnh nhân có tổng thời gian nằm viện nhỏ hơn tổng thời gian nằm ICU. Điều này có thể xảy ra vì một số bệnh nhân có tình trạng khẩn cấp phải vào điều trị ICU trước khi làm thủ tục nhập viện. Tuy nhiên, theo chúng tôi với những bệnh nhân có chênh lệch giữa tổng thời gian nằm viện và thời gian nằm ICU lần đầu lớn chúng cần bị loại bỏ, vì ta không rõ nguyên nhân gây ra hiện tượng đó. 

Từ hai lập luận bên trên, chúng tôi đề xuất một giải pháp dung hòa để xử lí vấn đề này như sau: loại bỏ ra khỏi tập dữ liệu những bệnh nhân có thời gian nằm ICU lần đầu lớn hơn tổng thời gian nằm viện với điều kiện khoảng cách giữa 2 khoảng thời gian này không vượt quá 0,5 ngày(điều kiện dung hòa). Khi đó, đối với những trường hợp bệnh nhân được giữ lại nhờ điều kiện dung hòa như trên, chúng tôi sẽ gán tổng thời gian nằm viện chính là thời gian nằm ICU lần đầu.

```python
# Lọc bỏ những bệnh nhân có tổng thời gian nằm viện nhỏ hơn thời gian nằm ICU
patients = patients[
    ((patients['los_icu'] < patients['total_los']) &   # 'los_icu' lớn hơn 'total_los'
    ((patients['los_icu'] - patients['total_los']) <= 0.5))  # khoảng cách không vượt quá 0.5
]
```

```python
# Chuyển các giá trị có total_los < los_icu thành los_icu
patients.loc[patients['total_los'] < patients['los_icu'], 'total_los'] = patients['los_icu']

patients.loc[patients['total_los'] == patients['los_icu'], 'admittime'] = patients['intime'] 

patients.loc[patients['total_los'] == patients['los_icu'], 'dischtime'] = patients['outtime'] 
```
